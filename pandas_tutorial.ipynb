pandas series

```python
import pandas as pd

pd_series_1= pd.Series([10,23,34],index=['maths','biology','cs'])

print(pd_series_1['maths'])

#ways of implementation

#from a list

s1=pd.Series([12,23,43])

#from a list with labels

s2=pd.Series([12,14,23],index=['physics','maths','chemistry'])

#from a dictionary

s3=pd.Series({"maths":23,'physics':34,'chemistry':45})


print(s1)
print(s2)
print(s3)
```

accessing the elements

```python
print(s2[0])
print(s2["maths"])
print(s2[0:2])
```

**working with a dataset**

```python
dataset=pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv")
print(dataset.head())
```

```python
print(dataset['sepal_length'].head())
```

```python
print(dataset.loc[0:15])
```

```python
print(dataset.describe())
```

```python
print(dataset.info())
```

```python
print(dataset.iloc[0])
```

```python
print(dataset[['sepal_length','sepal_width']])
```

```python
print(dataset[(dataset['sepal_length']>4) & (dataset['species']=='setosa')])
```

**Aggregation**

```python
dataset.groupby("species").mean()
```

```python
dataset.groupby("species")['sepal_length'].agg(['mean','min','max'])
```

**Sorting and joining**

```python
dataset.sort_values(by='sepal_length')
```
