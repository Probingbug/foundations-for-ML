```python
import numpy as np
import matplotlib.pyplot as plt
```

generate the simple dataset

```python
np.random.seed(42)
x=np.linspace(0,10,50)
true_m,true_c=2,1
y=true_m*x+true_c+np.random.normal(0,2,x.shape)
```

initialize parameters

```python
m,c=0.0,0.0
learning_rate=0.01
iterations=1000
```

store loss for plotting

```python
loss_history=[]
```

Gradient descent

```python
for i in range (iterations):
  #predicted y values
  y_pred=m*x+c

  #compute gradients
  m_grad=(-2/len(x))*np.sum((y-y_pred)*x)
  c_grad=(-2/len(x))*np.sum(y-y_pred)

  #update parameters

  m=m-learning_rate*m_grad
  c=c-learning_rate*c_grad

  # compute loss

  loss=np.mean((y-y_pred)**2)
  loss_history.append(loss)

  #print updates every 100 iterations

  if i%100==0:
    print(f"iterations {i} :loss= {loss:.4f},m={m:.4f},c={c:.4f}")




```

print final parameters

```python
print(f"final parameters : m {m : .4f}, c{ c: .4f}")
```

plot the datasets

```python
plt.figure(figsize=(10,6))
plt.scatter(x,y,label="original data",color="Blue")
plt.plot(x,y_pred,label="predicted line",color="Red")
plt.title("linear Regression")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

plot loss

```python
plt.figure(figsize=(10,6))
plt.plot(range(iterations),loss_history,label="loss")
plt.xlabel("iterations")
plt.ylabel("loss")
plt.title("loss vs iterations")
plt.legend()
plt.show()
```
